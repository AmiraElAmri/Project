```python
forecast_dates = pd.date_range(start=1/1/2025 + 1/1/2030, periods=num_days_5_years, freq='B') # 'B' for business days
```

```python
# Assuming 'forecast_dates' is a DatetimeIndex and 'forecast_ci' is a DataFrame with a column named 'lower Close' and 'upper Close'
plt.figure(figsize=(14, 7))
plt.plot(train['Date'], train['Close'], label='Training Data', color='blue')
plt.plot(test['Date'], test['Close'], label='Actual Prices (Test)', color='green')
plt.plot(test['Date'], test['Predicted_Close'], label='Predicted Prices', color='red')

# Plot the forecast mean and confidence intervals
plt.plot(forecast_dates, forecast_mean, label='5-Year Forecast', color='orange')
plt.plot(forecast_dates, forecast_ci.iloc[:, 0], label='Lower Confidence Interval', color='gray', linestyle='--') # Assuming lower confidence interval is in the first column
plt.plot(forecast_dates, forecast_ci.iloc[:, 1], label='Upper Confidence Interval', color='gray', linestyle='--') # Assuming upper confidence interval is in the second column


plt.legend(loc='upper left')
plt.title('Gold Price Prediction using ARIMA (2025-2030)')
plt.xlabel('Date')
plt.ylabel('Gold Price (Close)')

plt.show()
```
